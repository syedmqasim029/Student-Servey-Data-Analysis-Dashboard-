## 3. Data Cleaning

```python
print('=== BEFORE CLEANING ===')
print(f'Total rows: {len(df)}')
print(f'\nMissing values:')
print(df.isnull().sum())
print(f'\nExpected Salary stats (raw):')
print(df['Expected_Salary'].describe())
```

```python
# ── Salary outlier removal ──
# Realistic range for Pakistani students: 10,000 – 1,000,000 PKR/month
SALARY_MAX = 1_000_000
SALARY_MIN = 10_000

outliers = df[df['Expected_Salary'] > SALARY_MAX]
print(f'Salary outliers removed: {len(outliers)}')
print(outliers[['Gender','Major','Expected_Salary']])

df_clean = df[
    (df['Expected_Salary'] >= SALARY_MIN) &
    (df['Expected_Salary'] <= SALARY_MAX)
].copy()

print(f'\nRows after cleaning: {len(df_clean)}')
```

```python
# ── Data type checks & ordinal encoding ──

year_order = {'1st': 1, '2nd': 2, '3rd': 3, '4th': 4}
df_clean['Year_Num'] = df_clean['Year'].map(year_order)

support_order = {'Neutral': 0, 'Highly Supportive': 1}
df_clean['Support_Num'] = df_clean['Family_Support'].map(support_order)

mentor_order = {'No': 0, 'Searching': 1, 'Yes': 2}
df_clean['Mentor_Num'] = df_clean['Mentor_Availability'].map(mentor_order)

print('=== AFTER CLEANING ===')
print(df_clean.dtypes)
print(f'\nExpected Salary stats (cleaned):')
print(df_clean['Expected_Salary'].describe().round(2))
```

```python
# ── Summary statistics ──
numeric_cols = ['Career_Certainty','CGPA','Skill_Confidence','Job_Market_Hope',
                'Self_Learn_Hours','Expected_Salary']
summary = df_clean[numeric_cols].describe().round(3)
print('Summary Statistics:')
summary
```

## 4. Exploratory Data Analysis (EDA)

```python
# ── Value counts for categorical columns ──
for col in ['Gender','Major','Year','City_Type','Mentor_Availability','Family_Support','Biggest_Hurdle']:
    print(f'\n{col}:')
    print(df_clean[col].value_counts().to_string())
```

```python
# ── Correlation matrix ──
corr_cols = ['Career_Certainty','CGPA','Skill_Confidence','Job_Market_Hope',
             'Self_Learn_Hours','Support_Num','Mentor_Num']
corr = df_clean[corr_cols].corr().round(3)
print('Correlation Matrix:')
corr
```

## 5. Plotly Visualizations

### 5.1 Biggest Career Hurdles

```python
hurdle_counts = df_clean['Biggest_Hurdle'].value_counts().reset_index()
hurdle_counts.columns = ['Hurdle', 'Count']
hurdle_counts = hurdle_counts.sort_values('Count', ascending=True)

colors = ['#534AB7','#1D9E75','#D85A30','#D4537E','#378ADD','#BA7517']

fig1 = px.bar(
    hurdle_counts, x='Count', y='Hurdle',
    orientation='h',
    title='Biggest Career Hurdles Reported by Students',
    color='Hurdle',
    color_discrete_sequence=colors,
    text='Count'
)
fig1.update_traces(textposition='outside')
fig1.update_layout(
    showlegend=False,
    xaxis_title='Number of Students',
    yaxis_title='',
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Arial', size=13),
    title_font_size=16,
    height=400
)
fig1.update_xaxes(showgrid=True, gridcolor='#f0f0f0')
fig1.show()
```

### 5.2 Mentor Availability & Family Support

```python
fig2 = make_subplots(
    rows=1, cols=2,
    specs=[[{'type':'domain'}, {'type':'domain'}]],
    subplot_titles=['Mentor Availability', 'Family Support']
)

mentor_counts = df_clean['Mentor_Availability'].value_counts()
support_counts = df_clean['Family_Support'].value_counts()

fig2.add_trace(go.Pie(
    labels=mentor_counts.index,
    values=mentor_counts.values,
    hole=0.5,
    marker_colors=['#534AB7','#D85A30','#1D9E75'],
    textinfo='label+percent'
), row=1, col=1)

fig2.add_trace(go.Pie(
    labels=support_counts.index,
    values=support_counts.values,
    hole=0.5,
    marker_colors=['#1D9E75','#888780'],
    textinfo='label+percent'
), row=1, col=2)

fig2.update_layout(
    title_text='Mentor Availability vs Family Support',
    title_font_size=16,
    font=dict(family='Arial', size=12),
    height=380
)
fig2.show()
```

### 5.3 Gender & City Type Distribution

```python
fig3 = make_subplots(
    rows=1, cols=2,
    specs=[[{'type':'domain'}, {'type':'xy'}]],
    subplot_titles=['Gender Distribution', 'City Type of Respondents']
)

gender_counts = df_clean['Gender'].value_counts()
city_counts = df_clean['City_Type'].value_counts().reset_index()
city_counts.columns = ['City','Count']

fig3.add_trace(go.Pie(
    labels=gender_counts.index,
    values=gender_counts.values,
    marker_colors=['#378ADD','#D4537E','#888780'],
    textinfo='label+percent',
    hole=0.4
), row=1, col=1)

fig3.add_trace(go.Bar(
    x=city_counts['City'],
    y=city_counts['Count'],
    marker_color=['#534AB7','#D85A30','#1D9E75'],
    text=city_counts['Count'],
    textposition='outside'
), row=1, col=2)

fig3.update_layout(
    title_text='Gender & City Type of Survey Respondents',
    title_font_size=16,
    showlegend=False,
    font=dict(family='Arial', size=12),
    plot_bgcolor='white',
    paper_bgcolor='white',
    height=400
)
fig3.update_yaxes(showgrid=True, gridcolor='#f0f0f0', row=1, col=2)
fig3.show()
```

### 5.5 Confidence & Hope by University Year

```python
year_group = df_clean.groupby('Year')[['Skill_Confidence','Job_Market_Hope']].mean().round(2).reset_index()
year_order_list = ['1st','2nd','3rd','4th']
year_group['Year'] = pd.Categorical(year_group['Year'], categories=year_order_list, ordered=True)
year_group = year_group.sort_values('Year')

fig5 = go.Figure()
fig5.add_trace(go.Bar(
    name='Avg Skill Confidence',
    x=year_group['Year'],
    y=year_group['Skill_Confidence'],
    marker_color='#534AB7',
    text=year_group['Skill_Confidence'],
    textposition='outside'
))
fig5.add_trace(go.Bar(
    name='Avg Job Market Hope',
    x=year_group['Year'],
    y=year_group['Job_Market_Hope'],
    marker_color='#1D9E75',
    text=year_group['Job_Market_Hope'],
    textposition='outside'
))
fig5.update_layout(
    barmode='group',
    title='Average Confidence & Hope by University Year',
    title_font_size=16,
    xaxis_title='University Year',
    yaxis_title='Average Score (out of 10)',
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Arial', size=13),
    yaxis=dict(range=[0,12]),
    height=420
)
fig5.update_yaxes(showgrid=True, gridcolor='#f0f0f0')
fig5.show()
```

## 6. Key Findings Summary

```python
print('=' * 60)
print('KEY FINDINGS FROM SURVEY DATA')
print('=' * 60)

top_hurdle = df_clean['Biggest_Hurdle'].value_counts().idxmax()
top_hurdle_pct = df_clean['Biggest_Hurdle'].value_counts(normalize=True).max() * 100
print(f'\n1. Top hurdle: {top_hurdle} ({top_hurdle_pct:.0f}%)')

no_mentor = (df_clean['Mentor_Availability'] == 'No').mean() * 100
print(f'2. Students with NO mentor: {no_mentor:.0f}%')

avg_conf_hs = df_clean[df_clean['Family_Support']=='Highly Supportive']['Skill_Confidence'].mean()
avg_conf_n  = df_clean[df_clean['Family_Support']=='Neutral']['Skill_Confidence'].mean()
print(f'3. Avg confidence (Highly Supportive family): {avg_conf_hs:.2f}')
print(f'   Avg confidence (Neutral family):           {avg_conf_n:.2f}')

corr_cgpa_cert = df_clean['CGPA'].corr(df_clean['Career_Certainty'])
print(f'4. Correlation CGPA vs Career Certainty: {corr_cgpa_cert:.3f}')

corr_mentor_conf = df_clean['Mentor_Num'].corr(df_clean['Skill_Confidence'])
print(f'5. Correlation Mentor Access vs Skill Confidence: {corr_mentor_conf:.3f}')

low_self_learn = (df_clean['Self_Learn_Hours'] <= 2).mean() * 100
print(f'6. Students self-learning ≤2 hrs/day: {low_self_learn:.0f}%')

print('\n' + '=' * 60)
print('These findings align with:')
print('- Lairio & Penttinen (2006): Guidance gap in higher education')
print('- Onoshakpokaiye (2023): Parental support & career confidence')
print('- Sarwar et al. (2024): Self-knowledge & career decision-making')
print('- Clements & Kamau (2018): Goal commitment & proactive behaviours')
print('=' * 60)
```
